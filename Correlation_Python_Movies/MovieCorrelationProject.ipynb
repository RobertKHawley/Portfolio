```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8) # Adjusts size of plot figures we will generate


# Read data

df = pd.read_csv(r'C:\Users\Robert\Documents\PythonCorrelationProject\movies.csv')
```

```python
# Observe data

df.head()
```

```python
# Check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} = {:%}'.format(col, pct_missing))
```

```python
# Data types

df.dtypes
```

```python
# Create correct year column
df['year_correct'] = df['released'].astype(str).str[:4].astype('int64')

df
```

```python
# Sort by gross

df = df.sort_values(by=['gross'], inplace=False, ascending=False)

df
```

```python
# Delete duplicates

df.drop_duplicates()

df # There were no duplicates
```

```python
# Drop 'year' column

df.drop(columns=['year'], axis=1, inplace=True)

df
```

```python
# Plot budget vs. gross

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})

plt.title('Budget vs. Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
```

```python
# Pearson correlations (so far)

df.corr(method='pearson')
```

Create more numerical fields to analyse:

```python
# Count the categories that are less susceptible to selection bias (genre, rating)

df['genre_count'] = df.groupby('genre').genre.transform('count')
df['rating_count'] = df.groupby('rating').rating.transform('count')
```

```python
# Calculate gross profit (gross - budget)

df['net_profit'] = df['gross'] - df['budget']

col = 'budget'
pct_missing = np.mean(df[col] == 0)
print('Percentage of {} column that is zero (not provided): {:%}'.format(col, pct_missing))

df.loc[df['budget'] == 0, 'net_profit'] = None # Remove instances where budget is not provided
```

```python
# Calculate gross profit margin

df['net_profit_margin'] = df['net_profit'] / df['gross']
```

```python
# Plot net_profit vs. gross

sns.regplot(x='net_profit', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

Good fit and no outliers

```python
# Plot net_profit_margin vs. gross

sns.regplot(x='net_profit_margin', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

Poor fit and huge outliers (failed movies with large investments)

```python
# Remove outliers

# set negative values in net_profit_margin to None:
df.loc[df['net_profit_margin'] < 0, 'net_profit_margin'] = None

# Re-plot above
sns.regplot(x='net_profit_margin', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

Much better, no extreme values

```python
# Visualise correlations

correlation_matrix = df.corr(method='pearson')

# Plot matrix with custom colourmap
cmap = sns.diverging_palette(15, 135, s=100, l=20, as_cmap=True)
sns.heatmap(correlation_matrix, annot=True, center=0.0, vmin=-1.0, vmax=1.0, cmap=cmap)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Sort value pairs to interpret the strongest correlations

sorted_pairs = df.corr().unstack().sort_values(ascending=False)
```

```python
# Highest positive correlations

sorted_pairs[(sorted_pairs) > 0.4]
```

Ignore 1.0s as they are trivial

net_profit vs. gross is very strong: Large profit needs large revenue (quite trivial).

gross vs. budget is strong: Invest more money, make more money.

gross vs. votes is fairly strong: Films that make more money are voted higher / vice versa.

budget vs. votes is somewhat strong: Films which had higher investment were voted higher.

net_profit vs. votes is somewhat strong: Films which are voted higher make more profit / vice versa.

net_profit vs. net_profit_margin is significant: higher net profit means higher net profit margins (but some high profit films' net profits are small compared to their budgets).

```python
# Highest negative correlations

sorted_pairs[(sorted_pairs) < -0.25]
```

net_profit_margin vs. budget is significant: Films with higher investment will not 'win the lottery' so to speak. They may earn a large absolute profit but their high budget means that revenue will not be as high compared to that budget.
